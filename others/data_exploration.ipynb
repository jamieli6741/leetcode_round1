```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# load data
reg_data = pd.read_csv('../regression_data.csv')
cls_data = pd.read_csv('../classification_data.csv')
ages = reg_data['age'].tolist()
heights = reg_data['height'].tolist()

# data exploration
plt.scatter(ages, heights)
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()


# check correlation
correlation = np.corrcoef(ages, heights)[0,1]  # 0.528 < 0.7, knn regression
print(correlation)
```
